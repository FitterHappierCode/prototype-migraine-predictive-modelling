# Migraine Dataset Inspection

This notebook provides an initial inspection of the migraine dataset. The purpose is to understand the structure and quality of the data before any machine-learning model is developed.

## 1. Import pandas

Pandas is a widely used Python library for working with tabular data. It allows us to load, inspect, and summarise the dataset in a clear way.

```python
import pandas as pd
```

## 2. Load the dataset

The dataset is loaded from the `data` folder using a relative path from inside the `notebooks` folder.

```python
df = pd.read_csv('../data/migraine_dataset.csv')
```

## 3. Display the first five rows

Viewing the first five rows gives a quick overview of how the dataset is organised and what the values look like.

```python
df.head()
```

## 4. Show the dataset shape

The shape shows the number of rows and columns in the dataset.

```python
print('Dataset shape:', df.shape)
```

## 5. Show all column names

This step lists every column so that the available variables can be checked clearly.

```python
print('Column names:')
print(df.columns.tolist())
```

## 6. Check missing values

Missing values are important to identify because they may affect later preprocessing and model performance.

```python
print('Missing values per column:')
print(df.isnull().sum())
```

## 7. Check data types

Inspecting data types helps us understand whether each variable is numeric, categorical, or stored as text.

```python
print('Data types:')
print(df.dtypes)
```

## 8. Show the class balance for `migraine_occurrence`

This shows the number of observations in each target class. It is useful for understanding whether the dataset is balanced.

```python
print('Class balance for migraine_occurrence:')
print(df['migraine_occurrence'].value_counts())
```

## 9. Show the distribution for `migraine_severity`

This helps us understand how frequently each severity level appears in the dataset.

```python
print('Distribution of migraine_severity:')
print(df['migraine_severity'].value_counts())
```

## 10. Define the input features

The following self-reported lifestyle variables are selected as the input features for later machine-learning work.

```python
features = [
    'sleep_hours',
    'mood_level',
    'stress_level',
    'hydration_level',
    'screen_time'
]
```

## 11. Define the target variable

The target variable is the outcome that a future model will aim to predict.

```python
target = 'migraine_occurrence'
```

## 12. Create `X` and `y`

Here, `X` contains the selected input features and `y` contains the target variable.

```python
X = df[features]
y = df[target]
```

## 13. Print the shapes of `X` and `y`

This confirms the dimensions of the feature matrix and the target vector.

```python
print('Shape of X:', X.shape)
print('Shape of y:', y.shape)
```

## 14. Print the target balance again

Finally, the distribution of the target variable is checked again after defining `y`.

```python
print('Target balance in y:')
print(y.value_counts())
```
